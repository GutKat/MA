```python
from Bio import AlignIO

def fasta_to_stk(input_fasta, output_stk):
    """
    Converts a FASTA alignment file to Stockholm (STK) format.

    Parameters:
    input_fasta: str - Path to the input FASTA alignment file.
    output_stk: str - Path to the output Stockholm (STK) file.
    """
    # Read the FASTA alignment
    alignment = AlignIO.read(input_fasta, "fasta")
    
    # Write the alignment to Stockholm format
    #with open(output_stk, "w") as stk_file:
        AlignIO.write(alignment, stk_file, "stockholm")

    print(f"Converted {input_fasta} to {output_stk}")

# Example usage
input_fasta = "/Users/katringutenbrunner/Desktop/MA/working/cov_model/MBFV_xrRNA1/MBFV_xrRNA1.aln"
output_stk =  "/Users/katringutenbrunner/Desktop/MA/working/cov_model/MBFV_xrRNA1/MBFV_xrRNA1_aln.stk"
fasta_to_stk(input_fasta, output_stk)
```
