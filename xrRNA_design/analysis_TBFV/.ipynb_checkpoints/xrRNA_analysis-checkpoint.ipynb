## GC Content

```python
import os
#import csv
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_stockholm(file_name):
    with open(file_name, "r") as f:
        file_data = f.read()
    
    file_data = file_data.split("\n")
    seq = [line.split(" ") for line in file_data if re.match("\w",line)]
    seq = [[e for e in line if e] for line in seq]
    return seq

def calculate_gc_content(string):
    string = string.lower()
    gc_content = (string.count("g") + string.count("c")) / (string.count("g") + string.count("c") + string.count("a") + string.count("u"))
    return gc_content

```

```python
stockholm_folder = "/scr/aldea/skibinski/at_mtw/tbfv_stk"
stockholm_files = os.listdir(stockholm_folder)
stockholm_files = [stockholm_folder + "/" + file for file in stockholm_files]

data_for_df = []
names = []
for file in stockholm_files:
    name = file[37:48]
    if name[-1] == ".":
        name = name[:-1]
    names.append(names)
    if '1' in name:
        xrRNA = 1
    else:
        xrRNA = 2
    for id, seq in read_stockholm(file):
        data_for_df.append([id, seq, len(seq), calculate_gc_content(seq), xrRNA, name])

df_stats = pd.DataFrame(data_for_df, columns=['id', 'seq', 'len', 'gc', 'xrRNA', 'name'])
print(len(df_stats))
print(f"GC mean: {df_stats['gc'].mean()}")
print(f"length mean: {df_stats['len'].mean()}")
```

```python
grouped = df_stats.groupby('name')
gc_means = {}
for group_name, group in grouped:
    gc_means[group_name] = group['gc'].mean()

means_sorted = sorted(gc_means.items(), key=lambda x:x[1])
means_sorted = dict(means_sorted)
plt.bar(*zip(*means_sorted.items()))
plt.ylim(0.4, 0.7)
plt.margins(0.0002)
#plt.hlines(0.5, -1, 16, color='r')
plt.show()
```

## analyze the structures of the TBFV

```python
import os
import re
import pandas as pd
import numpy as np

stats= 'data/TBFV_ss.csv'
df = pd.read_csv(stats, skiprows=13, delimiter=";")
print(df)
```

```python
#only want xrRNA1
xrRNA1_df = df[df['xrRNA-no'] == 1]
#only want samples with both pseudoknots
xrRNA1_df_clean = xrRNA1_df[xrRNA1_df['PKs?'] == '1+2']
xrRNA1_df_clean = xrRNA1_df_clean.drop(columns=['PKs?', 'xrRNA-no', 'numSeqs'])

#only want samples with all 3 stems
xrRNA1_df_clean = xrRNA1_df_clean[xrRNA1_df_clean['ML-I/III'] == "-"]
xrRNA1_df_clean = xrRNA1_df_clean.drop(columns=['ML-I/III'])


#remove columns which have only 0s / -
col_no_values = xrRNA1_df_clean.columns[(xrRNA1_df_clean == "-").all()]
print("Columns have no data:", list(col_no_values))
xrRNA1_df_clean = xrRNA1_df_clean.drop(columns=col_no_values)

#remove columns which have only 0s / -
same_value_columns = xrRNA1_df_clean.columns[xrRNA1_df_clean.nunique() == 1]
print("Columns have only same value:")
for i in list(same_value_columns):
    print("\t", i, "=", xrRNA1_df_clean[i][0])
print("\n")
xrRNA1_df_clean = xrRNA1_df_clean.drop(columns=same_value_columns)

# change - to 0s
xrRNA1_df_clean = xrRNA1_df_clean.replace('-', 0)

# add Interloop 1 add up, (have some with 2 loops, add to one)
xrRNA1_df_clean = xrRNA1_df_clean.replace('2+2', 4)

temp_df = xrRNA1_df_clean.drop(columns='name').astype(int)
xrRNA1_df_clean['IB-I'] = temp_df['IL-I'] + temp_df['BL-I']

#double the nucleotides in Stems because we counted basepairs not nucleotides
xrRNA1_df_clean['ST-I'] = temp_df['ST-I'] * 2
xrRNA1_df_clean['ST-II'] = temp_df['ST-II'] * 2
xrRNA1_df_clean['ST-III'] = temp_df['ST-III'] * 2

print(xrRNA1_df_clean, '\n')
```

```python
temp_df = xrRNA1_df_clean.drop(columns='name').astype(int)
temp_df.describe()
```

```python
import matplotlib.pyplot as plt


def correlation_analysis(df, method = 'spearman', statistic_file = False, save_fig = False):
    stats_df = df.drop(columns='name').astype(int)
    #create some correlation analysis
    corr =  stats_df.corr(method=method, numeric_only=False)

    # Display the correlation matrix
    print("Pearson correlation matrix:")
    print(round(corr, 2))
    print(type(corr))
    if statistic_file:
        corr.to_csv(statistic_file)
    plt.figure(figsize=(10, 10))
    plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
    plt.colorbar(label=f'Correlation')
    plt.xticks(np.arange(len(corr)), corr.columns, rotation=45)
    plt.yticks(np.arange(len(corr)), corr.columns)
    plt.title(f'Correlation Heatmap by {method}')
    if save_fig:
        plt.savefig(save_fig)
    else:
        plt.show()
    return corr
```

```python
correlation_analysis(xrRNA1_df_clean, method="spearman")
```

```python
correlation_analysis(xrRNA1_df_clean, method="pearson")
```

```python
# add stem structure together for further analysis
xrRNA1_df_merge = xrRNA1_df_clean.copy()
temp_df = xrRNA1_df_clean.drop(columns='name').astype(int)
xrRNA1_df_merge['I'] = temp_df['ST-I'] + temp_df['IL-I'] + temp_df['BL-I']
xrRNA1_df_merge['II'] = temp_df['ST-II'] + temp_df['HL-II'] + temp_df['BL-II']
xrRNA1_df_merge['III'] = temp_df['ST-III'] + temp_df['HL-III']
xrRNA1_df_merge['HL'] = temp_df['HL-II'] + temp_df['HL-III']
xrRNA1_df_merge['sum'] = temp_df[list(temp_df.columns)].sum(axis=1)
xrRNA1_df_merge = xrRNA1_df_merge[['name', 'I', 'II', 'III', 'HL', 'sum']]
xrRNA1_df_merge
```

## relation calculations

important xrRNA1 of TBFV, exluding  XiFV, TYUV, SREV, MPFV since they to not have 'consensus' structure (do not have ML) <br>
all numbers are in nts (also the stems) <br>
important relation to calculate are:
<li>stem beta to hairpin loop beta = st2/hl2
<li>stem gamma to hairpin loop gamma = st3/hl2
<li>structure alpha to structure beta = sI/sII
<li>structure beta to structure gamma = sII/sIII
<li>structure alpha to structure ganna = sI/sIII



```python
all_data_df
```

```python
def calculate_relations(data_frame):
    st2_hl2 = []
    st3_hl3 = []
    sI_sII = []
    sII_sIII = []
    sI_sIII = []
    names = []
    for i in data_frame.index:
        st1 = data_frame.loc[i, "ST-I"]
        st2 = data_frame.loc[i, "ST-II"]
        st3 = data_frame.loc[i, "ST-III"]
        hl2 = data_frame.loc[i, "HL-II"]
        hl3 = data_frame.loc[i, "HL-III"]

        sI = data_frame.loc[i, "I"]
        sII = data_frame.loc[i, "II"]
        sIII = data_frame.loc[i, "III"]
        
        try:
            st2_hl2.append(float(st2)/float(hl2))
        except:
            st2_hl2.append(np.NaN)
        try:
            st3_hl3.append(float(st3)/float(hl3))
        except:
            st3_hl3.append(np.NaN)

        try:
            sI_sII.append(float(sI)/float(sII))
        except:
            sI_sII.append(np.NaN)

        try:
            sII_sIII.append(float(sII)/float(sIII))
        except:
            sII_sIII.append(np.NaN)

        try:
            sI_sIII.append(float(sI)/float(sIII))
        except:
            sI_sIII.append(np.NaN)


        names.append(data_frame.loc[i, "name"])
    relations = {"name": names, "STII/HLII": st2_hl2, "STIII/HLIII": st3_hl3, 'SI/SII': sI_sII, 'SII/SIII': sII_sIII, 'SI/SIII': sI_sIII, }
    df_relations = pd.DataFrame(relations)
    df_relations = df_relations.replace(0, np.NaN)
    return df_relations

```

```python
df_rel = calculate_relations(all_data_df)
df_rel
```

## length distribution

```python
import os
from Bio import AlignIO
import matplotlib.pyplot as plt
import numpy as np
```

```python
stk_folder = '/scr/aldea/kgutenbrunner/data/tbfv_stk/'
stk_files = [stk_folder + file for file in os.listdir(stk_folder) if file.endswith('xrRNA1.stk')]
```

```python
sequences = []
rfs = []
for stk_file in stk_files:
    align = AlignIO.read(stk_file, "stockholm")
    rfs.append(align.column_annotations['reference_annotation'])
    for record in align:
        sequences.append(str(record.seq))
```

```python
sequence_lengths =[len(seq) for seq in sequences]
plt.hist(sequence_lengths)
plt.title('Histogram of all sequence lengths')
plt.xlabel('length')
plt.ylabel('frequency')
```

```python
print(np.min(sequence_lengths))
print(np.max(sequence_lengths))
print(np.mean(sequence_lengths))
```

```python
sequence_lengths =[len(seq) for seq in rfs]
plt.hist(sequence_lengths)
plt.title('Histogram of reference annotation lengths')
plt.xlabel('length')
plt.ylabel('frequency')
```

```python
import RNA
```

```python
mfes = []
for sequence in sequences:
    fc = RNA.fold_compound(sequence)
    fc.pf()
    (ss, mfe) = fc.mfe()
    mfes.append(mfe)
mfes = np.array(mfes)
```

```python
print(np.min(mfes))
print(np.max(mfes))
print(np.mean(mfes))
```
